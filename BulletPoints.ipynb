```python
import re
```

```python
def cleanText(words):
    """The function to clean text"""
    words = re.sub("[^a-zA-Z]"," ",words)
    text = words.lower().split()
    return " ".join(text)

bull['BULLET_POINTS'] = bull['BULLET_POINTS'].apply(cleanText)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import sent_tokenize, word_tokenize

from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from sklearn.model_selection import train_test_split
```

```python
stop = stopwords.words('english')
lem = WordNetLemmatizer()
```

```python
def remove_stopwords(text):
    """The function to removing stopwords"""
    text = [word.lower() for word in text.split() if word.lower() not in stop]
    return " ".join(text)
```

```python
nltk.download('wordnet')
```

```python
def word_lem(text):
    """The function to apply lemmatizing"""
    lem_text = [lem.lemmatize(word) for word in text.split()]
    return " ".join(lem_text)
bull['BULLET_POINTS'] = bull['BULLET_POINTS'].apply(remove_stopwords)
bull['BULLET_POINTS'] = bull['BULLET_POINTS'].apply(word_lem)
```

```python
bull['BULLET_POINTS'] = bull['BULLET_POINTS'].astype(str)
print(bull['BULLET_POINTS'].apply(lambda x: len(x.split(' '))).sum())
```

```python
import string

# Create a function to remove punctuations
def remove_punctuation(sentence: str) -> str:
    return sentence.translate(str.maketrans('', '', string.punctuation))
```

```python
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
bull['BULLET_POINTS'] = bull['BULLET_POINTS'].apply(porter.stem)

bull['BULLET_POINTS'] = bull['BULLET_POINTS'].apply(remove_punctuation)
```

```python
from sklearn.feature_extraction.text import CountVectorizer as cv,TfidfVectorizer
# CountVectorizer - name & categories
cv = CountVectorizer(min_df=10)
X_name = cv.fit_transform(bull['BULLET_POINTS'])
```

```python
from collections import Counter
def reducecolumns(Col):
    # remove columns with <= 1% non-zero entries (arbitrary choice so that notebook doesn't crash)
    n_docs = Counter(Col.nonzero()[1])
    cols_to_keep = [k for k, v in n_docs.items() if v > .0005 * Col.shape[0]]
    return Col[:, cols_to_keep]
```

```python
print("Item Name Shape: " + str(reducecolumns(X_name).shape))
```

```python
# Perform a TFIDF Transformation of the item description with the top 55000 features and has an n-gram range of 1-2
tv = TfidfVectorizer(max_features=55000, ngram_range=(1, 2), stop_words='english')
X_description = tv.fit_transform(bull['BULLET_POINTS'])
```

```python
tfidf_weights = dict(zip(tv.get_feature_names_out(), tv.idf_))
tfidf_weights = pd.DataFrame(columns = ['BULLET_POINTS']).from_dict(dict(tfidf_weights), orient = 'index')
tfidf_weights.columns = ['BULLET_POINTS']

#Lowest tfidf scores -  Highest frequency words that get most importance:
tfidf_weights.sort_values(by=['BULLET_POINTS'], ascending=True).head(10)
```
